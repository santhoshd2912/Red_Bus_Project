```python
# Import necessary libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
from selenium.webdriver.common.action_chains import ActionChains
import time
import os
```

Andhara

```python
#Get Route name and link
df_AP=pd.read_csv(r"C:/Users/My_Redbus_project/Redbus_route_link/DF_AP.csv")
```

```python
df_AP
```

```python
#Extract  the AP bus details

Bus_details_AP=webdriver.Chrome()

Bus_names_AP      = []
Bus_types_AP      = []
start_time_AP     = []
End_time_AP       = []
Ratings_AP        = []
Total_duration_AP = []
Price_AP          = []
Seat_Available_AP = []
Route_names       = []
Route_links       = []


for i,r in df_AP.iterrows():
    link=r["Route_link"]
    route=r["Route_name"]
    
#loop through each link
    Bus_details_AP.get(link)
    time.sleep(5)

    
#elements to view each bus details:
    elements=Bus_details_AP.find_elements(By.XPATH,f"//a[contains(@href,'{link}')]") 
    for element in elements:
        element.click()
        time.sleep(3)


#view bus Details in Pages:
    try:
       next_page=Bus_details_AP.find_element(By.XPATH,"//div[@class='button']")
       next_page.click()

    except:
        continue
    time.sleep(2)

    scrolling = True
    while scrolling:
       old_page_source = Bus_details_AP.page_source
    # For Down the Page:
       ActionChains(Bus_details_AP).send_keys(Keys.PAGE_DOWN).perform()
       time.sleep(5)
       new_page_source = Bus_details_AP.page_source

       if new_page_source == old_page_source:
           scrolling = False


#Extract Bus Details

    bus_name      =Bus_details_AP.find_elements(By.XPATH,"//div[@class='travels lh-24 f-bold d-color']")
    bus_type      =Bus_details_AP.find_elements(By.XPATH,"//div[@class='bus-type f-12 m-top-16 l-color evBus']")
    start_time    =Bus_details_AP.find_elements(By.XPATH,"//*[@class='dp-time f-19 d-color f-bold']")
    end_time      =Bus_details_AP.find_elements(By.XPATH,"//*[@class='bp-time f-19 d-color disp-Inline']") 
    total_duration=Bus_details_AP.find_elements(By.XPATH,"//*[@class='dur l-color lh-24']")

    try:
        
       rating = Bus_details_AP.find_elements(By.XPATH,"//div[@class='clearfix row-one']/div[@class='column-six p-right-10 w-10 fl']")
    
    except:
         continue
        

    price =Bus_details_AP.find_elements(By.XPATH,'//*[@class="fare d-block"]')
    seats =Bus_details_AP.find_elements(By.XPATH,"//div[contains(@class,'seat-left')]")

#Add the data into created List

    for name in bus_name:
            Bus_names_AP.append(name.text)
            Route_names.append(route)
            Route_links.append(link)
    for duration in total_duration:
            Total_duration_AP.append(duration.text)
    for type in bus_type:
            Bus_types_AP.append(type.text)
    for s_time in start_time: 
            start_time_AP.append(s_time.text)
    for e_time in end_time:
            End_time_AP.append(e_time.text)
    for ratings in rating: 
            Ratings_AP.append(ratings.text)
    for travel_price in price:  
            Price_AP.append(travel_price.text)
    for s_available in seats:
            Seat_Available_AP.append(s_available.text)


print("Sucessfully completed")

    
```

```python
#Covert data into DataFrame

Bus_Details={
    "bus_name":Bus_names_AP,
    "total_duration":Total_duration_AP,
    "bus_type":Bus_types_AP,
    "start_time":start_time_AP,
    "end_time":End_time_AP,
    "rating":Ratings_AP,
    "price":Price_AP,
    "seats":Seat_Available_AP,
    "Route_name":Route_names,
    "Route_link":Route_links

    }

AP=pd.DataFrame(Bus_Details)
AP["state_name"]="Andhra Pradesh"
#convert Data into CSV 

AP.to_csv("AP.csv",index=False)
AP
```

```python
#Extract  the AP bus details

Bus_details_HR=webdriver.Chrome()

Bus_names_HR      = []
Bus_types_HR      = []
start_time_HR     = []
End_time_HR       = []
Ratings_HR        = []
Total_duration_HR = []
Price_HR          = []
Seat_Available_HR = []
Route_names       = []
Route_links       = []

for i,r in df_H.iterrows():
    link=r["Route_link"]
    route=r["Route_name"]

#loop through each link
    Bus_details_HR.get(link)
    time.sleep(5)
#elements to view each bus details:
    elements=Bus_details_HR.find_elements(By.XPATH,f"//a[contains(@href,'{link}')]") 
    for element in elements:
        element.click()
        time.sleep(3)


#view bus Details in Pages:
    try:
       next_page=Bus_details_HR.find_element(By.XPATH,"//div[@class='button']")
       next_page.click()

    except:
        continue
    time.sleep(2)

    scrolling = True
    while scrolling:
       old_page_source = Bus_details_HR.page_source
    # For Down the Page:
       ActionChains(Bus_details_HR).send_keys(Keys.PAGE_DOWN).perform()
       time.sleep(5)
       new_page_source = Bus_details_HR.page_source

       if new_page_source == old_page_source:
           scrolling = False


#Extract Bus Details
    bus_name      =Bus_details_HR.find_elements(By.XPATH,"//div[@class='travels lh-24 f-bold d-color']")
    bus_type      =Bus_details_HR.find_elements(By.XPATH,"//div[@class='bus-type f-12 m-top-16 l-color evBus']")
    start_time    =Bus_details_HR.find_elements(By.XPATH,"//*[@class='dp-time f-19 d-color f-bold']")
    end_time      =Bus_details_HR.find_elements(By.XPATH,"//*[@class='bp-time f-19 d-color disp-Inline']") 
    total_duration=Bus_details_HR.find_elements(By.XPATH,"//*[@class='dur l-color lh-24']")

    try:
        
       rating = Bus_details_HR.find_elements(By.XPATH,"//div[@class='clearfix row-one']/div[@class='column-six p-right-10 w-10 fl']")
    
    except:
         continue
        

    price =Bus_details_HR.find_elements(By.XPATH,'//*[@class="fare d-block"]')
    seats =Bus_details_HR.find_elements(By.XPATH,"//div[contains(@class,'seat-left')]")

#Add the data into created List

    for name in bus_name:
            Bus_names_HR.append(name.text)
            Route_names.append(route)
            Route_links.append(link)
    for duration in total_duration:
            Total_duration_HR.append(duration.text)
    for type in bus_type:
            Bus_types_HR.append(type.text)
    for s_time in start_time: 
            start_time_HR.append(s_time.text)
    for e_time in end_time:
            End_time_HR.append(e_time.text)
    for ratings in rating: 
            Ratings_HR.append(ratings.text)
    for travel_price in price:  
            Price_HR.append(travel_price.text)
    for s_available in seats:
            Seat_Available_HR.append(s_available.text)


print("Sucessfully completed")

    
```

```python
#Covert data into DataFrame

Bus_Details_HR={
    "bus_name":Bus_names_HR,
    "total_duration":Total_duration_HR,
    "bus_type":Bus_types_HR,
    "start_time":start_time_HR,
    "end_time":End_time_HR,
    "rating":Ratings_HR,
    "price":Price_HR,
    "seats":Seat_Available_HR,
    "Route_name":Route_names,
    "Route_link":Route_links
    }

HR=pd.DataFrame(Bus_Details_HR)
HR["state_name"]="Himachal Pradesh"
HR.to_csv("df_HR.csv",index=False)
HR
```
