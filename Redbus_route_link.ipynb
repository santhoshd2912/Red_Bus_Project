#State wise Extract the Bus link and route 

```python
#UP
driver=webdriver.Chrome()
driver.get("https://www.redbus.in/online-booking/uttar-pradesh-state-road-transport-corporation-upsrtc")
time.sleep(5)
driver.maximize_window()
all_data=driver.page_source
```

```python
UP= WebDriverWait(driver,50)

def UP_link_route(path):
    UP_link = []
    UP_route= []

    # To be get Link:
    for i in range(1, 3):
        paths = driver.find_elements(By.XPATH,path)

        for links  in paths:
            link=links.get_attribute("href")
            UP_link.append(link)

        # To be get name of route:
        for route in paths:
            UP_route.append(route.text)

        try:
            pagination = UP.until(EC.presence_of_element_located((By.XPATH,'//*[@class="DC_117_paginationTable"]')))
            next_button = pagination.find_element(By.XPATH,f'//div[@class="DC_117_pageTabs " and text()={i + 1}]')
            time.sleep(10)
            next_button.click()
            
        except:
            print(f"No more pages to paginate at step")
             

    return UP_link,UP_route

UP_link,UP_route = UP_link_route('//a[@class="route"]')
```

```python
DF_UP=pd.DataFrame({"Route_name":UP_route,"Route_link":UP_link})
DF_UP
```

```python
#convert Data frame into CSV formet

DF_UP.to_csv("DF_UP.csv",index=False)
```
